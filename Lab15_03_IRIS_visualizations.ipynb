```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.FacetGrid(iris,hue='Species',height=4).map(plt.scatter, 'PetalLengthCm', 'PetalWidthCm').add_legend()
```

```python
sns.FacetGrid(iris,hue='Species',height=4).map(sns.kdeplot, 'SepalLengthCm').add_legend()
```

```python
sns.boxplot(x='Species',y='PetalLengthCm',data=iris)
```

```python
sns.stripplot(x='Species',y='SepalLengthCm',data=iris)
```

```python
sns.violinplot(x='Species',y='PetalLengthCm',data=iris)
```

```python
sns.pairplot(iris.drop('Id',axis=1),hue='Species')
```
